### Задача 1
Ковариация и корреляция. Выполните следующие операции:  

1. Сгенерируйте набор данных:

- `X = np.random.normal(5, 2, 100)`
- `Y = 2 * X + np.random.normal(0, 1, 100)`

2. Вычислите ковариацию и корреляцию между X и Y.

3. Постройте график зависимости X-Y.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.normal(5, 2, 100)
Y = 2 * X + np.random.normal(0, 1, 100)

cov_matrix = np.cov(X, Y)
covariance = cov_matrix[0, 1]
correlation = np.corrcoef(X, Y)[0, 1]

plt.scatter(X, Y)
plt.title('График зависимости X-Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()

print('Ковариация между X и Y:', covariance)
print('Корреляция между X и Y:', correlation)
```

### Задача 2
Создайте две случайные матрицы размером 3х3 с числами из интервала [0, 10] с использованием NumPy.

Выполните следующие операции:  

1. Найдите их сумму и произведение.

2. Найдите определитель одной из матриц.

3. Найдите обратную матрицу (если существует).

```python
import numpy as np

np.random.seed(42) 
A = np.random.randint(0, 11, (3, 3))
B = np.random.randint(0, 11, (3, 3))

AB_sum = A + B
AB_product = np.dot(A, B)
det_A = np.linalg.det(A)

try:
    inverse_A = np.linalg.inv(A)
except np.linalg.LinAlgError:
    inverse_A = None

print('Матрица A:\n', A)
print('Матрица B:\n', B)
print('Сумма матриц:\n', AB_sum)
print('Произведение матриц:\n', AB_product)
print('Определитель матрицы A:', det_A)
if inverse_A is not None:
    print('Обратная матрица A:\n', inverse_A)
else:
    print('Обратной матрицы для A не существует.')
```

### Задача 3
Визуализация функции потерь (логистическая регрессия). Выполните следующие операции:

1. Реализуйте логистическую функцию (сигмоиду): `sigma(x) =  1/(1 + exp(-x))`.

2. Сгенерируйте значения x и визуализируйте sigma(x) с помощью Matplotlib.

3. Реализуйте и визуализируйте функцию бинарной кросс-энтропии (логарифмической потери).


```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15 
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -y_true * np.log(y_pred) - (1 - y_true) * np.log(1 - y_pred)

x_values = np.linspace(-10, 10, 500)
sigmoid_values = sigmoid(x_values)

y_true = 1 
y_pred_values = np.linspace(0.01, 0.99, 500)  
loss_values = binary_cross_entropy(y_true, y_pred_values)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x_values, sigmoid_values)
plt.title('Сигмоида')
plt.xlabel('x')
plt.ylabel('sigma(x)')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(y_pred_values, loss_values, color='red')
plt.title('Функция бинарной кросс-энтропии')
plt.xlabel('Предсказанная вероятность')
plt.ylabel('Потеря (Loss)')
plt.grid()

plt.show()
```

### Задача 4
Работа с матрицами и векторами (линейная алгебра). Выполните следующие операции:  

1. Умножьте две матрицы:  

- `A = [[3, 2], [1, 4]]`,

- `B = [[2, 0], [1, 6]]`.

2. Найдите нормы данного вектора: `v = [3, 4, 5]`. 

3. Вычислите определитель и обратную матрицу A.

```python
import numpy as np

A = np.array([[3, 2], [1, 4]])
B = np.array([[2, 0], [1, 6]])
v = np.array([3, 4, 5])

AB_product = np.dot(A, B)

vector_norm_1 = np.linalg.norm(v, ord=1)
vector_norm_2 = np.linalg.norm(v, ord=2)
vector_norm_inf = np.linalg.norm(v, ord=np.inf)

det_A = np.linalg.det(A)
try:
    inv_A = np.linalg.inv(A)
except np.linalg.LinAlgError:
    inv_A = "Обратной матрицы не существует."

print('Произведение матриц A и B:\n', AB_product)
print('Норма L1:', vector_norm_1) 
print('Норма L2:', vector_norm_2)
print('Норма бесконечности:', vector_norm_inf)
print('Определитель матрицы А:', det_A)
print('Обратная матрица A:\n', inv_A)
```

### Задача 5
Реализация линейной регрессии (с использованием градиентного спуска), используя библиотеку numpy. Используйте этот набор данных:

- `х = [1, 2, 3, 4, 5]`

- `y = [2, 4, 6, 8, 10]`

Выполните следующие операции:

1. Реализуйте линейную регрессию `y = wx + b`, используя градиентный спуск.  

2. Найдите значения w и b, которые минимизируют ошибку MSE.  

3. Визуализируйте входные данные и линию регрессии.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
w = 0.0 
b = 0.0 
learning_rate = 0.01
num_iterations = 1000

def gradient_descent(x, y, w, b, learning_rate, num_iterations):
    n = len(x)
    for i in range(num_iterations):
        y_pred = w * x + b  
        dw = (-2/n) * np.sum(x * (y - y_pred))  
        db = (-2/n) * np.sum(y - y_pred) 
        w = w - learning_rate * dw  
        b = b - learning_rate * db 
        if i % 100 == 0:
            mse = np.mean((y - y_pred) ** 2)
            print(f"Iteration {i}: w = {w}, b = {b}, MSE = {mse}")
    return w, b

w, b = gradient_descent(x, y, w, b, learning_rate, num_iterations)

plt.scatter(x, y, color='blue', label='Исходные данные')
plt.plot(x, w * x + b, color='red', label='Линия регрессии')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Линейная регрессия с использованием градиентного спуска')
plt.legend()
plt.show()

print(f"Оптимальные значения: w = {w}, b = {b}")
```
