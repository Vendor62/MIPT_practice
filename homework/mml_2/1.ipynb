```python
import json

notebook_path = "mml_2.ipynb"
output_path = "mml_2.ipynb"

with open(notebook_path, "r", encoding="utf-8") as f:
    nb = json.load(f)

def remove_widgets(obj):
    """Рекурсивно удаляет все ключи 'widgets' в любом месте"""
    if isinstance(obj, dict):
        if "widgets" in obj:
            del obj["widgets"]
        for v in obj.values():
            remove_widgets(v)
    elif isinstance(obj, list):
        for item in obj:
            remove_widgets(item)

remove_widgets(nb)

# Сохраняем исправленный ноутбук
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(nb, f, ensure_ascii=False, indent=1)

print(f"Исправленный ноутбук сохранён как {output_path}")
```
