3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_exp(string):
    """Функция определяет опыт работы в месяцах из входной строки

    Args:
        string (str): cтрока, содержащая информацию о опыте работы

    Returns:
        exp (int): опыт работы в месяцах
    """
    years = 0
    months = 0
    year_keys = ['год', 'года', 'лет']
    month_keys = ['месяц', 'месяцев', 'месяца']
    
    if string is np.nan or string == 'Не указано':
        return np.nan
    
    lst = string.split(' ')[:6]
    
    for i, key in enumerate(lst):
        if key in year_keys:
            years += int(lst[i-1])
        
        if key in month_keys:
            months += int(lst[i-1])
            
    exp = months + years * 12
    
    return exp
    
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(get_exp)
df = df.drop('Опыт работы', axis=1)

print('Медианный опыт работы в месяцах:', df['Опыт работы (месяц)'].median())
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_city(string):
    """Функция определяет город соискателя из входной строки

    Args:
        string (str): cтрока, содержащая информацию о городе соискателя

    Returns:
        str: город соискателя
    """
    million_cities = [
        'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 
        'Казань', 'Челябинск', 'Омск', 'Самара', 
        'Ростов-на-Дону', 'Уфа', 'Красноярск', 
        'Пермь', 'Воронеж', 'Волгоград' 
    ]
    lst = string.split(' ')
    
    if lst[0] == 'Москва':
        return 'Москва'
    elif lst[0] == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif lst[0] in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'
    
def get_removal(string):
    """Функция определяет готовность соискателя к переезду

    Args:
        string (str): строка, содержащая информацию о готовности к переезду

    Returns:
        bool: True или False
    """ 
    lst = string.split(',')
    
    if 'м.' in lst[1]:
        key = 2
    else:
        key = 1
    
    if 'не' in lst[key]:
        return False
    else:
        return True

def get_command(string):
    """Функция определяет готовность соискателя к коммандировкам

    Args:
        string (str): строка, содержащая информацию о готовности к командировкам

    Returns:
        bool: True или False
    """
    lst = string.split(',')
    
    if 'м.' in lst[1]:
        key = 3
    else:
        key = 2
    
    if 'ком' in string:
        if 'не' in lst[key]:
            return False
        else:
            return True
    else:
        return False   

df['Город'] = df['Город, переезд, командировки'].apply(get_city)
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(get_removal)
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(get_command)
df = df.drop('Город, переезд, командировки', axis=1)

print(
    'Доля соискателей из Санкт-Петербурга:', 
    round(
        ((df[df['Город'] == 'Санкт-Петербург'].count() / len(df)) * 100).values[0]
    )
)
print(
    'Доля готовых и к переезду и к командировкам:',
    round(
        ((df['Готовность к переезду'] & df['Готовность к командировкам']).sum() / len(df)) * 100
    )
)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg><br>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
employments = [
    'полная занятость', 'частичная занятость', 
    'проектная работа', 'волонтерство', 'стажировка'
]
schedules = [
    'полный день', 'сменный график', 
    'гибкий график', 'удаленная работа', 
    'вахтовый метод'
]

for i in employments:
    df[i] = df['Занятость'].apply(lambda x: i in x)
    
for i in schedules:
    df[i] = df['График'].apply(lambda x: i in x)
    
df = df.drop(['Занятость', 'График'], axis=1)
    
print(
    'Ищут и проектную работу и волонтёрство:', 
    (df['проектная работа'] & df['волонтерство']).sum()
)
print(
    'Хотят работать вахтовым методом и с гибким графиком:', 
    (df['вахтовый метод'] & df['гибкий график']).sum()
)
```

```python
rates = pd.read_csv("D:\\ds\\mipt\\ExchangeRates.csv")

df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме']).dt.date
rates['date'] = pd.to_datetime(rates['date']).dt.date

df['ЗП(руб)'] = df['ЗП'].apply(lambda x: int(x.split(' ')[0]))
df['Валюта'] = df['ЗП'].apply(lambda x: x.split(' ')[1])

replaces = {
    'руб.':'RUB', 'грн.':'UAH', 'USD':'USD', 
    'EUR':'EUR', 'бел.руб.':'BYN', 'KGS':'KGS', 
    'сум':'UZS', 'AZN':'AZN', 'KZT':'KZT'
}
df['Валюта'] = df['Валюта'].map(replaces)

df_merged = df.merge(
    rates, 
    how='left', 
    left_on=['Валюта', 'Обновление резюме'],
    right_on=['currency', 'date']
)
df_merged[['close', 'proportion']] = df_merged.apply(
    lambda row: pd.Series({
        'close': 1 if row['Валюта'] == 'RUB' else row['close'],
        'proportion': 1 if row['Валюта'] == 'RUB' else row['proportion']
    }),
    axis=1
)
df_merged['ЗП(руб)'] = df_merged['ЗП(руб)'] * df_merged['close'] / df_merged['proportion']

df = df_merged.drop(
    [
        'Пол, возраст', 'ЗП', 'Образование и ВУЗ', 
        'Валюта', 'currency', 'per', 'date', 'time', 
        'close', 'vol', 'proportion'
    ], axis=1
)

print('Медианная зарплата составляет:', df_merged['ЗП(руб)'].median())
df.info()
```

6. В результате анализа мы обнаружили потенциальные выбросы в признаке **"Возраст"**. Это оказались резюме людей чересчур преклонного возраста для поиска работы. Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
log_age = np.log(df['Возраст'])

mean_age = log_age.mean()
std_age = log_age.std()

lower_bound = mean_age - 3 * std_age
upper_bound = mean_age + 4 * std_age  

plt.figure(figsize=(10, 6))
histplot = sns.histplot(
    log_age, 
    kde=True, 
    color='blue', 
    bins=30
)

histplot.axvline(
    mean_age, 
    color='g', 
    lw=2, 
    label='Среднее'
)
histplot.axvline(
    lower_bound, 
    color='r', 
    linestyle='--', 
    lw=2, 
    label='Нижняя граница (3 сигмы)'
)
histplot.axvline(
    upper_bound, 
    color='r', 
    linestyle='--', 
    lw=2, 
    label='Верхняя граница (4 сигмы вправо)'
)

plt.title('Распределение возраста (логарифмическое масштабирование)')
plt.xlabel('Возраст (логарифм)')
plt.ylabel('Частота')
plt.legend()
plt.grid(True)
plt.show()

skewness = log_age.skew()
print(f"Асимметрия распределения (skewness): {skewness:.3f}")

def outliers_z_score_mod(data, feature, left=3, right=4, log_scale=False):
    """
    Функция для поиска выбросов в данных с использованием метода z-отклонений.
    
    Args:
    ----------
    data : pandas.DataFrame
        Таблица с данными, в которой содержится анализируемый признак.
    feature : str
        Название столбца (признака) для поиска выбросов.
    left : float, optional, default=3
        Количество стандартных отклонений влево от среднего значения для границы выбросов.
    right : float, optional, default=4
        Количество стандартных отклонений вправо от среднего значения для границы выбросов.
    log_scale : bool, optional, default=False
        Флаг, указывающий на необходимость использования логарифмического масштаба для анализа.
        Если True, анализируется логарифмированное значение признака (log(x + 1)).

    Returns:
    ----------
    outliers : pandas.DataFrame
        Таблица с выбросами, которые находятся за пределами заданных границ.
    cleaned : pandas.DataFrame
        Таблица с очищенными данными, содержащая только записи внутри заданных границ.
    """
    if log_scale:
        x = np.log(data[feature] + 1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned


outliers, cleaned_data = outliers_z_score_mod(
    df, 
    'Возраст', 
    left=3, 
    right=4, 
    log_scale=True
)
print(f"Количество выбросов: {outliers.shape[0]}")
display(outliers)
print(
    'Возраст соискателей, попавших под категорию выбросов:', 
    outliers['Возраст'].values
)
```
