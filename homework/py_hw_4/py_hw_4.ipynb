### Задание №1

#### Найти пересечение без дубликатов

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом 'A' и клиентом 'B' на платформе «Авито»).

Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.

**Важно!** Если пересечение пустое, то на экран должен выводиться пустой список.

Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу `Некорректный ввод`.

Для поиска пересечений в списках напишите функцию `pure_intersection()`, которая будет возвращать список с уникальными элементами пересечения.


**Пример №1:**
```
Ввод:
108, 138, 42, 52, 14
109, 13, 52, 32, 42, 14, 109

Вывод:
[42, 52, 14]

```

**Пример №2:**
```
Ввод:
108, 138, 42, 52, 14
189, 202, 249, 24, 15, 3

Вывод:
[]

```

**Пример №3:**
```
Ввод:

189, 202, 249, 24, 15, 3

Вывод:
Некорректный ввод

```

**Подсказка:**

Для определения пересечения преобразуйте списки во множества и воспользуйтесь методом `intersection()`.

````python
input_string1 = input('Введите 1-ую последовательность идентификаторов: ')
input_string2 = input('Введите 2-ую последовательность идентификаторов: ')
def pure_intersection(string1, string2):
    """
    Данная функция:
    1. Разделяет каждую из двух сторк на подстроки по запятой и преобразует эти подстроки в целые числа, создавая два списка.
    2. Преобразует эти списки в множества и находит их пересечение. Функция работает только с двумя строками, поэтому оператор &
       предпочтительнее для определения пересечений, чем метод intersection(), т.к. позволяет реализовать короткую запись.
    3. Преобразует пересечения обратно в список и возвращает его.
    4. Если в п.1 не удаётся создать список из целых чисел, то возвращает сообщение об ошибке.
    """
    try:
        return list(set(map(int, string1.split(','))) & set(map(int, string2.split(','))))
    except ValueError:
        return 'Некорректный ввод'

print(pure_intersection(input_string1, input_string2))
````

## Задание №2

#### Найти минимум и максимум

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь подаёт на вход программе действительные числа через пробел с клавиатуры (через функцию `input()`).

Он привык пользоваться Excel, поэтому отделяет дробную часть числа запятой, а не точкой. Кроме того, он может случайно ввести число «словами» (например, не "5", а «пять») — в этом случае слово надо игнорировать.

**Ваша задача** — вывести минимум и максимум последовательности. Гарантируется, что пользователем будет введено хотя бы одно число.

Для нахождения минимума и максимума напишите функцию `find_min_max`, которая будет возвращать два числа.

**Пример:**

```
Ввод:
-2,56 25,002 восемь 19,13 13 -37,5

Вывод:
Minimum: -37.5
Maximim: 25.002
```



**Подсказка №1:**

Для проверки того, состоит ли строка `s` только из букв, можно воспользоваться методом `s.isalpha()`.

**Подсказка №2:**

Для поиска минимума и максимума в списке можно воспользоваться встроенными функциями `min()` и `max()`.

````python
input_string = input('Введите последовательность чисел: ')
def find_min_max(string):
    """
    Данная функция:
    1. Заменяет во вводимой последовательности запятые на точки.
    2. Разбивает строку на подстроки по пробелам и проверяет состоит ли подстрока из букв. Если да, то такая подстрока игнорируется.
    3. Преобразует подстроки, прошедшие фильтр, в вещественные числа и создаёт из них итоговый список.
    4. Возвращает строку, содержащую минимальное и максимальное числа из этого списка.
    """
    string = string.replace(',', '.')
    lst = [float(x) for x in string.split() if not x.isalpha()]
    return f'Minimum: {min(lst)}\nMaximum: {max(lst)}'

print(find_min_max(input_string))
````

## Задание №3

#### Найти медиану

**Компания:** «Авито»

**Уровень:** Лёгкий

---
Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

---
Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел неупорядоченную последовательность чисел.

Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.

Для поиска медианы в списке напишите функцию `find_median()`, которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.

Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза `Некорректный ввод`.

**Пример №1:**
```
Ввод:
1, 5, 2, 3, 6

Вывод:
Median: 3.0
```
```
Ввод:
100, 5, 2, 4, 3, 6

Вывод:
Median: 4.5
```
```
Ввод:

Вывод:
Некорректный ввод
```
```
Ввод:
десять, 10, пять, 7, семь

Вывод:
Некорректный ввод
```


````python
input_string = input('Введите последовательность чисел: ')
def find_median(string):
    """
    Данная функция:
    1. Разбивает вводимую последовательность на подстроки по запятой и формирует из них упорядоченный по возрастанию список,
       элементами в котором являются целые числа.
    2. Определяет длину списка.
    3. Проверяет является ли длина чётным числом, используя остаток от деления на 2.
    4. Если длина нечётная, то находит индекс среднего элемента при помощи целочисленного деления на 2.
    5. Если длина чётная, то считает среднее арифметическое от двух центральных элементов.
    6. Возвращает число, являющееся медианой.
    7. Если в п.1 не удаётся создать список из целых чисел, то возвращает сообщение об ошибке.
    """
    try:
        lst = sorted(map(int, string.split(',')))
        n = len(lst)
        if n % 2 != 0:
            return f'Median: {lst[n // 2]}'
        else:
            return f'Median: {(lst[n // 2 - 1] + lst[n // 2]) / 2}'
    except ValueError:
        return 'Некорректный ввод'

print(find_median(input_string))
````

**Подсказка №2:**

При решении задачи вам может пригодиться метод строк `s.startswith(word)`, который определяет, начинается ли строка `s` со строки `word`. Подробнее о нём вы можете узнать в [документации](https://docs-python.ru/tutorial/operatsii-tekstovymi-strokami-str-python/metod-str-startswith/).

````python
input_string = input('Введите число словами: ')
def transform_string_to_integer(string):
    """
    Логика такая:
    1. Если на ввод передаётся, например, "семьсот семьдесят семь тысяч семьсот семьдесят семь", то после поиска соответствий
       в словаре, получим набор "700 70 7 1000 700 70 7". Складываем то, что слева от 1000, умножаем на 1000 и добавляем сумму
       чисел справа от тысячи.
    2. Если "1000 700 70 7", то тысячу добавляем в результат напрямую, а затем складываем то, что справа от тысячи.
    3. Если на ввод передаются сотни или десятки, то просто складываем их.
    """
    number_word_dict = {
        "ты": 1000, "м": 1000000, "сто": 100, "двес": 200, "трис": 300, "четырес": 400,
        "пятьс": 500, "шестьс": 600, "семьс": 700, "восемьс": 800, "девятьс": 900,
        "одинн": 11, "двен": 12, "трин": 13, "четырн": 14, "пятн": 15, "шестн": 16,
        "семн": 17, "восемн": 18, "девятн": 19, "двад": 20, "трид": 30, "сор": 40,
        "пятьд": 50, "шестьд": 60, "семьд": 70, "восемьд": 80, "девяно": 90, "дес": 10,
        "н": 0, "о": 1, "дв": 2, "т": 3, "ч": 4, "п": 5, "ш": 6, "с": 7, "в": 8, "д": 9
    }
    word_lst = string.split()  # формирую список с отдельными словами
    result_num = 0  # сюда буду складывать итоговое число
    current_sum = 0  # здесь будет временная сумма
    for word in word_lst:  # иду по отдельным словам в списке
        for key in number_word_dict:  # вложенный цикл идёт по ключам в словаре
            if word.startswith(key):  # ищу совпадение первых букв в отдельном слове с ключом в словаре
                num = number_word_dict[key]  # когда нашёл совпадение, то по ключу забираю число из словаря в переменную
                if num == 1000: # если встречаю тысячу
                    if current_sum == 0: # то проверяю не пуста ли временная сумма
                        result_num += num  # если временная сумма пуста, то добавляю число напрямую
                    else: # а если нет
                        result_num += current_sum * num # то умножаю её на тысячу
                    current_sum = 0 # обнуляю временную сумму, чтоб корректно посчитать оставшиеся порядки
                else:  # если нахожу число меньше тысячи
                    current_sum += num  # складываю его во временную сумму
                break  # перебирающий ключи цикл прервётся при первом совпадении
    result_num += current_sum  # после завершения циклов временная сумма добавляется к итоговому числу
    return result_num

print(transform_string_to_integer(input_string))
````

### Задание №5 (бонусная, не оценивается)

#### Разложить факториал на простые множители

**Компания:** «СБЕР»

**Уровень:** Сложный

___

Пользователь с клавиатуры (через функцию `input()`) вводит некоторое целое число `n`.

Напишите функцию `decompose_factorial()`, которая будет принимать на вход введённое пользователем число `n`, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.

---

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

Например, факториал числа $3$ будет равен: $3! = 1 * 2 * 3 = 6$.

---

**Простое число** — число, которое делится (нацело) только на себя и на 1. Например, $13$ — простое число, так как оно делится только на себя (на $13$) и на $1$, а $14$ не является простым числом, так как оно делится на $7$ и на $2$.

Подробнее о простых числах можно почитать [здесь](https://dzen.ru/a/XqFMe_VWclT8NTss).

---
**Простые множители** — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа $10$ простыми множителями являются числа $5$ и $2$. Для числа $28$ множителями являются числа $7$ и $4$, но простыми множителями будут числа $7$ и $2$, так как $4$ — это $2 * 2$, а число $28$ представляется в виде $28 = 7 * 2^2$.

----

В результате своей работы функция должна возвращать строку вида (приведён пример для `n=12`):

`12! = 2^10 * 3^5 * 5^2 * 7 * 11`

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

**Пример №1:**
```
Ввод:
5

Вывод:
5! = 2^3 * 3 * 5
```

**Пример №2:**
```
Ввод:
22

Вывод:
22! = 2^19 * 3^9 * 5^4 * 7^3 * 11^2 * 13 * 17 * 19
```

**Пример №3:**
```
Ввод:
25

Вывод:
25! = 2^22 * 3^10 * 5^6 * 7^3 * 11^2 * 13 * 17 * 19 * 23

```



**Подсказка №1:**

**Подсказка №2:**

Алгоритм нахождения простых множителей числа вы можете найти [здесь](https://foxford.ru/wiki/informatika/razlozhenie-chisla-na-mnozhiteli-v-python).

````python
n = int(input('Введите число: '))

def decompose_factorial(n):
    """
    Поскольку функция для нахождения простых множителей уже дана в условии задачи, то вкладываю её в основную функцию без изменений.
    Остаётся лишь циклом пройтись по каждому числу в факториале и разложить его на простые множители, посчитать сколько раз множитель
    встречается в разложении и вывести это на экран в требуемом формате. В данном решении мне не нужна функция factorial. Как вариант,
    можно было бы находить функцией factorial значение факториала, затем функцией factor раскладывать его на множители и затем
    подсчитывать количество вхождений каждого элемента в список, но вариант с циклом представляется мне более предпочтительным.
    """
    def factor(n):
        ans = []
        d = 2
        while d * d <= n:
            if n % d == 0:
                ans.append(d)
                n //= d
            else:
                d += 1
        if n > 1:
            ans.append(n)
        return ans
    if n == 0 or n == 1:
        return f'{n}! = 1'
    else:
        factor_dict = {} # в этом словаре буду считать сколько раз встречается каждый множитель
        for i in range(2, n + 1): # цикл пойдёт по каждому числу в факториале
             factor_lst = factor(i) # каждое число будет разложено на простые множители, которые будут записаны в список
             for j in factor_lst: # цикл пойдёт по множителям в спике
                 if j in factor_dict: # если множитель уже есть в списке
                     factor_dict[j] += 1 # то увеличит его счётчик на 1
                 else: # а если встречается впервые
                     factor_dict[j] = 1 # то счётчик устанавливается на единицу
        result = [] # в этом списке данные будут подготавливаться к выводу на экран
        for num, count in factor_dict.items(): # цикл будет перебирать множители и значения счётчика
            if count == 1: # если множитель встречается только раз
                result.append(f"{num}") # то добавляем его в результат как есть
            else: # а если нет
                result.append(f"{num}^{count}") # то в результат пойдёт число и значение счётчика в указанном формате
        output = ' * '.join(result) # объединяю элементы списка с указанным разделителем
        return f'{n}! = {output}' # строка подаётся на вывод в нужном формате

print(decompose_factorial(n))
````

## Задание №6 (бонусная, не оценивается)

#### Заменить дублирующиеся буквы

**Компания:** «СБЕР»

**Уровень:** Сложный

Пользователь с клавиатуры (через функцию `input()`) вводит строку, состоящую из символов латинского алфавита.

Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.

**Несколько важных моментов:**

* буквы не обязаны быть смежными (смежные — стоящие рядом; в строке `'abc'` буквы a и b, а также b и c являются смежными, в то время как a и c — нет);
* zz заменяем на a;
* вводимая строка состоит из букв латинского алфавита в нижнем регистре.

Напишите функцию `replace_duplicates()`, которая принимает на вход произвольную строку `s`, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

**Пример:**
```
input:
zzzab

step 1: zaba
step 2: zbb
step 3: zc

output:
cz
```

```
input:
hhakafh

step 1: haakfi
step 2: hkfib

output:
hkfib
```

**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string

s = input('Введите строку: ')

def replace_duplicates(s): # вроде работает, но это не точно
    """
    Логика такая:
    1. Функция делает один проход по строке (списку) и если находит две одинаковых буквы, то получает их индексы и удаляет.
    2. Добавляет букву для замены.
    3. Переключает флаг, отмечая, что сделала проход и замену.
    4. Рекурсивно заходит на вторую проходку, выполняя п 1, 2 и 3, если снова находит одинаковые буквы.
    5. Если одинаковых букв больше нет, то флаг остаётся в положении False и функция выходит из рекурсии.
    6. Возвращает обработанную строку.
    """

    def get_next_char(letter): # это функция для нахождения следующей буквы для замены
        alphabet = string.ascii_lowercase # тут лежит алфавит
        index = alphabet.index(letter) # тут определяется индекс буквы, которую надо заменить
        next_index = (index + 1) % len(alphabet) # а тут индекс буквы на которую нужно заменить, если 26, то будет 0
        return alphabet[next_index] # возвращаем буквы для замены

    def process(s): # эта фунция делает всю основную работу
        s_lst = list(s) # разбиваю строку в список с отдельными буквами
        count_dict = {} # словарь для подсчета
        for i in s_lst: # цикл идёт по буквам в списке
            if i in count_dict: # если буква уже есть в списке
                count_dict[i] += 1 # то увеличит его счётчик на 1
            else: # а если встречается впервые
                count_dict[i] = 1 # то счётчик устанавливается на единицу
        replacement_flag = False # здесь флаг, который отслеживает была ли произведена замена
        for letter_in_dict, count in count_dict.items(): # цикл идёт буквам и их счётчику в словаре-счётчике
            if count > 1: # если буква встречается более 1 раза, то:
                indices = []  # сюда буду складывать индексы таких букв при помощи цикла
                for index, letter_in_list in enumerate(s_lst): # цикл идёт по буквам и индексам в списке с исходной строкой
                    if letter_in_list == letter_in_dict: # когда находим повторяющуюся букву в списке
                        indices.append(index) # то записываем её индекс в список
                        indices = indices[:2] # за одну проходку нужно сделать только одну замену, поэтому обрезаю список
                indices.sort(reverse=True) # сортирую список от большего к меньшему, чтоб корректно удалять
                for index in indices: # это цикл для удаления
                    del s_lst[index] # удаляю по одной первые две встреченные повторяющиеся буквы
                s_lst.append(get_next_char(letter_in_dict)) # и добавляю вместо них заменяющую букву
                replacement_flag = True # переключаю флаг
                break # и выхожу из цикла

        if replacement_flag: # если флаг в True
            return process(''.join(s_lst)) # то снова запускается поиск и замена
        else: # а если False
            return ''.join(s_lst) # то выхожу из рекурсии и возвращаю результат

    return process(s) # здесь рекурсивно вызывается функция обработки

print(replace_duplicates(s))
````
