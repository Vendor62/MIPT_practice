```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return torch.sin(x / 5) * torch.exp(x / 10) + 5 * torch.exp(-x / 5)

x = torch.tensor([15.0], requires_grad=True)  
learning_rate = 0.1
iterations = 100
history = []

for i in range(iterations):
    if x.grad is not None:
        x.grad.zero_()
    
    y = f(x)
    y.backward()
    
    with torch.no_grad():
        x -= learning_rate * x.grad
    
    history.append((x.item(), y.item()))

print(f"Минимум найден в x = {x.item():.4f}, f(x) = {f(x).item():.4f}")

x_vals = np.linspace(0, 30, 100)
y_vals = f(torch.tensor(x_vals)).detach().numpy()
plt.plot(x_vals, y_vals, label='Функция')
plt.scatter(*zip(*history), c='red', label='Шаги градиентного спуска')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
